```python
import gensim.models as g
```

```python
path = "enwiki_dbow/doc2vec.bin"
```

```python
doc2vec = g.Doc2Vec.load(path)
```

```python
import json, glob
```

```python
files = glob.glob("fashion-iq-master/captions/*")
```

```python
files[2]
```

```python
with open(files[1]) as json_file:
    data = json.load(json_file)
```

```python
from tqdm import tqdm
import numpy as np
```

```python
for f in files:
    
    with open(f) as json_file:
        data = json.load(json_file)

    for i in tqdm(data):
        embeds = []
        for c in i['captions']:
            embeds.append(doc2vec.infer_vector(c.split(" ")))
        i['caption_embeds'] = np.array(embeds).tolist()
    
    with open(f.split("/")[-1], 'w') as outfile:
        json.dump(data, outfile)
```
